#### Задание 1.

Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

`Precision` _(точность)_ - точность отражает какой процент объектов, отнесённых к "основному" классу, действительно являются таковыми.  
`Recall` _(полнота)_ - полнота отражает какой процент объектов "основного" класса было обнаружено алгоритмом.

1. При медицинском тестировании _False Negative_ будет давать ложно обнадеживающее сообщение пациентам и врачам о том, что болезнь отсутствует, когда в действительности она присутствует. В этом случае лучше максимизировать **Recall** получив, тем самым, большее количество ложных сообщений о наличии болезни _False Positive_, чтобы перепроверить данные и убедиться в её осутствии - это лучше чем пропустить болезнь.
2. При оценке на сейсмоактивности местности лучше максимизировать **Recall**, чтобы причислив место к сейсмоактивным и не пропустить действительно опасную местность.
3. Для фильтрации спама _False Positive_ возникает, когда фильтры спама или методы блокировки спама ошибочно классифицируют порядочное письмо как спам и, как следствие, препятствуют его прочтению, попаданию во «Входящие». Таким образом, _False negative_ здесь будет уместней чем _False Positive_, а значит в этом случае лучше максимизировать **Precision**.
4. Когда кредитная политика банка направлена на увеличение максимального количества платёжеспособных клиентов, а не на общее увеличение количества клиентов банка, то тогда имеет смысл максимизировать **Precision**, увеличив точность распознавания платёжеспособных клиентов.

#### Задание 2.

Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

$F-мера$ (в общем случае $F_\beta$) — среднее гармоническое Precision и Recall.  
$F_\beta = (1+\beta) * ((Precision*Recall) : ((\beta^2 * Precision) + Recall))$  
$\beta$ в данном случае определяет вес точности в метрике, и при $\beta = 1$ это среднее гармоническое (с множителем 2, чтобы в случае precision = 1 и recall = 1 иметь $\ F_1 = 1$)  
$F-мера$ достигает максимума при полноте (Recall) и точности (Precision), равными единице, и близка к нулю, если один из аргументов близок к нулю.

1. F-мера позволяет оценить Precision и Recall в комплексе.
2. F-мера позволяет увеличивать/уменьшать влияние Precision и Recall.
